## Training A Model(Training loop):

in this code first i create a tiny neural netrwork and then give it some random data and ask the model to print the result and it give me junk values and then i train the model for 100 times and then again ask model to print the results and this time model gives me some useful results as we can see in output.

this process includes some functions that are given below with their simple explanation:


1------  nn.Linear() : made a simple neural network(model)

2-----criterion=nn.CrossEntropyLoss(): measures diffrence between actual result and bot guess


3----optimizer=optim.Adam(): act as a mechanic and repair the faulty weights


4----model.parameters(): tells the mechanic which weights are faulty and you are allowed to touch
phase

# LOOP:

5----y_pred=model(input_data):tells the mind of bot to take this input and give me best possible guess

6----loss=criterion(y_pred,correct_class): calculates the error between guess and correct result.


7----loss.backword(): detects which wires(weights) are faulty and causing wrong guess

8---optimizer.step():act as a mechanic  optimize the faulty and resolve the issue

9----optimizer.zero_grad(): it clears the models memory so that it does not become messy

10---loss.item():value extraction, extract a float value for loss of data during training and print it nicely






```python
import torch
import torch.nn as nn
import torch.optim as optim


model = nn.Linear(3, 2)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.1)




input_data = torch.tensor([[1.0, 1.0, 1.0]])
correct_class = torch.tensor([0])

print("--- BEFORE TRAINING ---")
pred = model(input_data)
print(f"Prediction scores: {pred.tolist()}")



print("\n--- TRAINING start (Loop 100 times) ---")

for epoch in range(100):

    y_pred = model(input_data)


    loss = criterion(y_pred, correct_class)


    loss.backward()


    optimizer.step()


    optimizer.zero_grad()

    if epoch % 20 == 0:
        print(f"Epoch {epoch}: Loss error is {loss.item():.4f}")

print("\n")
pred = model(input_data)
print(f"Final Prediction scores: {pred.tolist()}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim



X_toppings = torch.tensor([[1.0], [2.0], [3.0], [4.0]])



Y_prices   = torch.tensor([[7.0], [9.0], [11.0], [13.0]])



model = nn.Linear(1, 1)



criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

print(f"Guess for 5 toppings BEFORE training: ${model(torch.tensor([5.0])).item():.2f}")


print("\n--- Training start for  Cashier ---")
for epoch in range(1000):


    pred_price = model(X_toppings)


    loss = criterion(pred_price, Y_prices)


    loss.backward()


    optimizer.step()


    optimizer.zero_grad()

    if (epoch+1) % 200 == 0:
        print(f"Epoch {epoch+1}: Error is {loss.item():.4f}")




print("\n")
guess = model(torch.tensor([5.0])).item()
print(f"Guess for 5 toppings AFTER training: ${guess:.2f}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

x = torch.tensor([[1.0, 2.0]])
y = torch.tensor([[4.0]])


model = nn.Linear(2, 1)

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.1)

print("--- BEFORE TRAINING ---")
print("Prediction:", model(x).item())



for epoch in range(100):

    pred = model(x)
    loss = criterion(pred, y)

    loss.backward()
    optimizer.step()
    optimizer.zero_grad()

    if epoch % 20 == 0:
        print(f"Epoch {epoch}, Loss = {loss.item():.4f}")

print("\n--- AFTER TRAINING ---")
print("Prediction:", model(x).item())
```
